```python
#G
extractedDf = dataset.loc[550:900, 'Gender':'Coughing of Blood']

extractedDf.to_csv('extracted_sample.csv', index=False)

print(f"Mean values:\n{extractedDf.mean()}\n")
print(f"Median values:\n{extractedDf.median()}\n")
print(f"Mode values:\n{extractedDf.mode().iloc[0]}\n")
print(f"Standard Deviation values:\n{extractedDf.std()}\n")
print(f"Minimum values:\n{extractedDf.min()}\n")
print(f"Maximum values:\n{extractedDf.max()}\n")
```

```python
#@title Task 03

#A

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.histplot(dataset['Age'], kde=True, color='blue')
plt.title('Distribution of Age')

plt.subplot(2, 2, 2)
sns.countplot(x='Gender', data=dataset, palette='Set2')
plt.title('Distribution of Gender')

plt.subplot(2, 2, 3)
sns.boxplot(y='Obesity', data=dataset, color='orange')
plt.title('Boxplot of Obesity')

plt.subplot(2, 2, 4)
sns.scatterplot(x='Alcohol use', y='Obesity', data=dataset, hue='Gender', palette='Set1')
plt.title('Alcohol use vs Obesity')

plt.tight_layout()
plt.show()
```

```python
#B

numericDataset = dataset.select_dtypes(include=[np.number])

correlation_matrix = numericDataset.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap of the Numeric Features')
plt.show()
```

```python
#C

plt.figure(figsize=(8, 6))
sns.jointplot(x='Age', y='Frequent Cold', data=dataset, kind='kde', fill=True, cmap='Blues')
plt.title('Distribution of Age and Frequent Cold', pad=70)
plt.show()
```

```python
#D

plt.figure(figsize=(6, 8))
sns.boxplot(y='Obesity', data=dataset, color='red')
plt.title('Boxplot to Identify Outliers in Obesity')
plt.show()
```
